```python
import numpy as np
import tensorflow as tf
import gym
```

```python
env = gym.make('FrozenLake-v0')
```

```python
input_state = tf.placeholder(tf.float32,[1,16])
W,B = tf.Variable(tf.random_uniform([16,4],0,0.02)),tf.Variable(0.2)
all_Q_values =tf.add(tf.matmul(input_state,W),B)
q_arg_max = tf.argmax(all_Q_values,1)
```

```python
def state_gen(idx):
    state = np.zeros([1,16])
    state[0,idx] = 1
    return state
```

```python
Q_next = tf.placeholder(tf.float32,[1,4])
cost = tf.reduce_sum(tf.square(Q_next - all_Q_values))
opt = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)
```

```python
policy = []
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    reward_list = []
    gamma = 0.99
    num_episode = 2000
    for i in xrange(num_episode):
        s = env.reset()
        d = False
        step_reward = 0
        for _ in xrange(100):
            Q,a = sess.run([all_Q_values,q_arg_max],feed_dict={input_state:state_gen(s)})
            #adding randomness
            if np.random.uniform(0,1) < 0.1 and i < 7000:
                a[0] = env.action_space.sample()
            #taking action
            s1,r,d,_ = env.step(a[0])
            if(s1 == 5 or s1 == 7 or s1 == 11 or s1 == 12):
                r = -2
            Q1 = sess.run([all_Q_values],feed_dict={input_state:state_gen(s)})
            Qmax = np.amax(Q1)
            Q[0,a[0]] = r + gamma * Qmax
            _,w = sess.run([opt,W],feed_dict={input_state:state_gen(s),Q_next:Q})
            s = s1
            step_reward += r
            if d == True:
                break
        reward_list.append(step_reward)
    
    # display policy    
    for itr in xrange(env.observation_space.n):
        policy.append(sess.run([all_Q_values],feed_dict={input_state:state_gen(itr)}))
```

```python
i=0
for ele in policy:
    print 'state: ' + str(i) + '\t value:\t' + str(ele[0].argmax())
    i+=1
```

```python
rewd = 0
for ele in reward_list:
    if ele > 0:
        rewd += ele
print rewd/2000.0
```
